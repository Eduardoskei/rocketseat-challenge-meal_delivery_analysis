##  Criação de Novas Colunas (Feature Engineering)

```python
df_orders['Receita_Item'] = df_orders['Quantidade'] * df_orders['Preco_Unitario']
```

```python
df_orders.head()
```

## Tratamento de Valores Ausentes

```python
# Identificando valores ausentes no df_orders
df_orders.isna().sum()
```

```python
# Identificando valores ausentes no df_menu
df_menu.isna().sum()
```

```python
# Preenchendo valores ausentes na coluna Quantidade com a média
df_orders['Quantidade'].fillna(df_orders['Quantidade'].mean(), inplace=True)
df_orders.isna().sum()
```

```python
# Removendo linhas com valores nulos na coluna Preco_Unitario
df_orders.dropna(subset=['Preco_Unitario'], inplace=True)
df_orders.isna().sum()
```

## Agregações por Item

```python
df_grouped = df_orders.groupby('Item').agg({
    'Quantidade': 'sum',
    'Receita_Item': 'sum'
})

df_grouped
```

```python
top_5_quantidade = df_grouped.sort_values(by='Quantidade', ascending=False)
top_5_quantidade.head()
```

```python
top_5_receita = df_grouped.sort_values(by='Receita_Item', ascending=False)
top_5_receita.head()
```

## Análise Temporal

```python
df_orders['Data'] = pd.to_datetime(df_orders['Data'])
```

```python
df_orders.info()
```

```python
df_orders['Mes'] = df_orders['Data'].dt.month
```

```python
receita_mensal = df_orders.groupby('Mes')['Receita_Item'].sum()
receita_mensal
```

## Integração de Dados (Merge)

```python
df_merged = pd.merge(df_orders, df_menu, on='Item', how='left')
df_merged.head()
```

```python
receita_categoria = df_merged.groupby('Categoria')['Receita_Item'].sum()
receita_categoria
```

```python
top_categoria = receita_categoria.sort_values(ascending=False)
top_categoria.head(1)
```

## Filtros e Consultas

```python
filtered = df_merged[(df_merged['Categoria'] == 'Salgados') & (df_merged['Quantidade'] > 10)]

filtered
```

## KPIs e Análise Estatística com NumPy

```python
receita_total = np.sum(df_merged['Receita_Item'])
receita_total
```
